```python
import numpy as np
from minisom import MiniSom

characters = {
    "G": [G,G_2,G_3,G_4,G_5,G_6],
    "N": [N,N_2,N_3,N_4,N_5,N_6],
    "I": [I,I_2,I_3,I_4,I_5,I_6],
    "Q": [Q,Q_2,Q_3,Q_4,Q_5,Q_6],
    "O": [O,O_2,O_3,O_4,O_5,O_6],
    "U": [U,U_2,U_3,U_4,U_5,U_6],
    "H": [H,H_2,H_3,H_4,H_5,H_6],
    "Z": [Z,Z_2,Z_3,Z_4,Z_5,Z_6]
}
data = np.array([representation for ch in "GNIQOUHZ" for representation in characters[ch]])
# print(data.shape)

row, col = 5, 5  # 二维拓扑结构
initial_eta = 0.6  # 初始学习率
final_eta = 0.01  # 最终学习率
num_iterations = 1000 

input_dim = 35
output_dim =25
w = np.random.rand(output_dim, input_dim) # weights matrix

def wta(x, w):
    distances = np.linalg.norm(w - x, axis=1)  # 计算距离
    return np.argmin(distances)  # 返回距离最小的索引

# 无拓扑结构更新
def compete0(x, w, eta):
    for xx in x:
        id = wta(xx, w)
        w[id] = w[id] + eta * (xx - w[id])
    return w

# 一维邻域查找
def neighborid1(id, length=25):
    neighbors = [id]
    if id > 0:          neighbors.append(id - 1)  # 左邻居
    if id < length - 1: neighbors.append(id + 1)  # 右邻居
    return neighbors

# 一维拓扑结构更新
def compete1(x, w, eta, length=25):
    for xx in x:
        id = wta(xx, w)
        neighbors = neighborid1(id, length)
        for i in neighbors:
            w[i] += eta * (xx - w[i])  # 更新邻域内的权重
    return w

# 二维邻域查找
def neighborid2(id, row, col):
    rown, coln = id // col, id % col
    neighbors = [id]
    if coln > 0:        neighbors.append(id - 1)  # 左邻居
    if coln < col - 1:  neighbors.append(id + 1)  # 右邻居
    if rown > 0:        neighbors.append(id - col)  # 上邻居
    if rown < row - 1:  neighbors.append(id + col)  # 下邻居
    return neighbors

# 二维拓扑结构更新
def compete2(x, w, eta, row=5, col=5):
    for xx in x:
        id = wta(xx, w)
        neighbors = neighborid2(id, row, col)
        for i in neighbors:
            w[i] += eta * (xx - w[i])  # 更新邻域内的权重
    return w
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 映射每个输入数据到竞争层中的胜出节点
def map_data_to_nodes(data, w):
    mappings = {}
    for idx, x in enumerate(data):
        winner_node = wta(x, w)
        mappings[idx] = winner_node
    return mappings

# 可视化函数：在2D平面上展示映射结果
def plot_mappings(mappings, title, row=None, col=None):
    plt.figure(figsize=(8, 6))
    if row and col:
        plt.title(f"{title} (Grid: {row}x{col})")
        plt.scatter([v % col for v in mappings.values()], [v // col for v in mappings.values()])
        for idx, node in mappings.items():
            plt.text(node % col, node // col, f"{list(characters.keys())[idx//6]}", ha='center', va='center')
    else:
        plt.title(f"{title} (Linear)")
        plt.scatter(mappings.values(), [0] * len(mappings))
        for idx, node in mappings.items():
            plt.text(node, 0, f"{list(characters.keys())[idx//6]}", ha='center', va='center')
    plt.show()

```

#### 无拓扑结构
在无拓扑结构下，竞争层中的每个节点相互独立，模型的更新只作用于最接近输入向量的胜出节点，而不会影响其他节点。
* 分散的节点分布：字符映射到竞争曾不同节点，而相似的字符可能没有聚集到一起，反映其缺乏邻域约束
* 部分字符集中映射而没有明显聚类
* 部分节点没有任何字符映射，无拓扑结构下，SOM权重更新更集中，没有充分利用整个竞争层。

```python
# 结果展示：无拓扑结构
print("Results for No Topology:")
w_0 = np.random.rand(output_dim, input_dim)  # 初始化权重
for i in range(num_iterations):
    eta = initial_eta - i * (initial_eta - final_eta) / num_iterations
    w_0 = compete0(data, w_0, eta)
mappings_0 = map_data_to_nodes(data, w_0)
print("Mappings (No Topology):", mappings_0)

# 可视化结果
plot_mappings(mappings_0, "Mappings for No Topology")
```

#### 一维拓扑结构
* 线性分布效果：相似字符在一维线性拓扑更容易映射到相邻或接近的位置。
* 由于更新时胜出节点的邻域节点也会更新，某些字符会映射在一起，其空间输入中比较相近。

```python
# 结果展示：一维拓扑结构
print("\nResults for 1D Topology:")
w_1d = np.random.rand(output_dim, input_dim)  # 初始化权重
for i in range(num_iterations):
    eta = initial_eta - i * (initial_eta - final_eta) / num_iterations
    w_1d = compete1(data, w_1d, eta)
mappings_1d = map_data_to_nodes(data, w_1d)
print("Mappings (1D Topology):", mappings_1d)

plot_mappings(mappings_1d, "Mappings for 1D Topology")
```

#### 二维拓扑结构
* 空间分布上，可以很好保持输入数据相似性关系
* 聚类效果：部分空间近似节点表现在网格中较近，而其余相距较远

```python
# 结果展示：二维拓扑结构
print("\nResults for 2D Topology:")
w_2d = np.random.rand(output_dim, input_dim)  # 初始化权重
for i in range(num_iterations):
    eta = initial_eta - i * (initial_eta - final_eta) / num_iterations
    w_2d = compete2(data, w_2d, eta, row, col)
mappings_2d = map_data_to_nodes(data, w_2d)
print("Mappings (2D Topology):", mappings_2d)

plot_mappings(mappings_2d, "Mappings for 2D Topology", 5, 5)
```

```python
# 动图生成类接口
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

class ImageSequenceGIF:
    def __init__(self, gif_dir='temp_gif_frames', gif_file='output.gif'):
        self.gif_dir = gif_dir
        self.gif_file = gif_file
        self.frames = []
        self.frame_count = 0

        if not os.path.exists(gif_dir):
            os.makedirs(gif_dir)

    def add_frame_from_plt(self, plt_figure=None):
        if plt_figure is None:
            plt_figure = plt.gcf()

        filename = os.path.join(self.gif_dir, f"frame_{self.frame_count:04d}.png")
        plt_figure.savefig(filename, bbox_inches='tight')
        self.frames.append(Image.open(filename))
        self.frame_count += 1
        plt.close(plt_figure)

    def save_gif(self, duration=100, last_frame_duration=200):
        if not self.frames:
            print("No frames to save.")
            return

        durations = [duration] * (len(self.frames) - 1) + [last_frame_duration]
        self.frames[0].save(
            self.gif_file,
            save_all=True,
            append_images=self.frames[1:],
            duration=durations,
            loop=0
        )
        print(f"GIF saved as {self.gif_file}")

    def clear_temp_frames(self):
        for f in os.listdir(self.gif_dir):
            os.remove(os.path.join(self.gif_dir, f))
        self.frames = []
        self.frame_count = 0
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 数据准备函数
def gif2_data(datanum=1000):
    x = np.random.rand(datanum, 2)

    condition1 = x[:, 1] <= x[:, 0]
    condition2 = x[:, 1] <= (1 - x[:, 0])
    index1 = np.where(condition1 & condition2)
    x1 = x[index1]
    x = np.delete(x, index1, axis=0)

    index2 = np.where(x[:, 0] <= 0.5)
    x2 = x[index2]
    x3 = np.delete(x, index2, axis=0)

    return np.vstack((x1, x2, x3))

# 选择数据集合
data = gif2_data(1000)

# 参数设置
input_dim = data.shape[1]  # 输入数据的维度
learning_rate = 0.5
epochs = 1000

# 定义用于显示训练结果的函数
def plot_som_result(data, w, topology, title):
    plt.figure(figsize=(8, 8))
    for i, x in enumerate(data):
        closest_node = wta(x, w)
        if topology == '1D':
            plt.plot(closest_node + np.random.rand() * 0.1, 0, 'bo')
        elif topology == '2D':
            row = closest_node // w.shape[1]
            col = closest_node % w.shape[1]
            plt.plot(col + np.random.rand() * 0.1, row + np.random.rand() * 0.1, 'ro')
    plt.title(f'SOM - {title}')
    plt.xlabel("Neuron Index" if topology == '1D' else "Column Index")
    plt.ylabel("" if topology == '1D' else "Row Index")
    plt.show()

# 训练并展示不同神经元数量和拓扑结构的影响
neuron_configs = {
    '1D - 10 Neurons': (10, '1D', compete1),
    '1D - 20 Neurons': (20, '1D', compete1),
    '2D - 5x5 Grid': ((5, 5), '2D', compete2),
    '2D - 10x10 Grid': ((10, 10), '2D', compete2)
}

for title, (neurons, topology, compete_fn) in neuron_configs.items():
    if topology == '1D':
        w = np.random.rand(neurons, input_dim)
        for epoch in range(epochs):
            eta = learning_rate * (1 - epoch / epochs)
            w = compete_fn(data, w, eta, length=neurons)
    elif topology == '2D':
        w = np.random.rand(neurons[0] * neurons[1], input_dim)
        for epoch in range(epochs):
            eta = learning_rate * (1 - epoch / epochs)
            w = compete_fn(data, w, eta, row=neurons[0], col=neurons[1])
    
    # 可视化训练结果
    plot_som_result(data, w, topology, title)
```

### Discussion

#### 神经元个数对训练结果的影响：
* 少量：SOM分辨能力较低；数据被粗略分类，相似度较低的数据也坑被分到一个或相邻的神经元中，聚类模糊。
* 较多：分辨能力提升；神经元可以专注于部分相似数据点，提高聚类精度。

#### 拓扑结构对训练结果影响：
* 一维：适合处理线性分布的数据；计算复杂度较低；对高维分布的数据被强制映射到一条线上，聚类效果粗糙。
* 二维：更好捕捉空间特征，聚类效果更好，计算复杂度较高。
