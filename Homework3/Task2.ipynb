```python
characters = {
    "G": G,
    "N": N,
    "I": I,
    "Q": Q,
    "O": O,
    "U": U,
    "H": H,
    "Z": Z,
}
data = np.array([characters[ch] for ch in "GNIQOUHZ"] * 6)

input_dim = 35
output_dim =25
w = np.random.rand(output_dim, input_dim) # weights matrix

def wta(data, w, epochs=1000):
    for epoch in range(epochs):
        for x in data:
            y = np.dot(w, x)
            winner = np.argmax(y)
            w[winner] += 0.1 * (x - w[winner])
    return w

# 无拓扑结构更新
def compete0(x, w, eta):
    for xx in x:
        id = wta(xx, w)
        w[id] = w[id] + eta * (xx - w[id])
    return w

# 二维邻域查找
def neighborid2(id, row, col):
    rown = id // col
    coln = id % col
    iddim = [id]

    if coln > 0:        iddim.append(id-1)
    if coln < col-1:    iddim.append(id+1)
    if rown > 0:        iddim.append(id-col)
    if rown < row-1:    iddim.append(id+col)

    return iddim

# 二维拓扑结构更新
def compete2(x, w, eta):
    for xx in x:
        id = wta(xx, w)
        iddim = neighborid2(id, 5, 5)

        for iidd in iddim:
            w[iidd] = w[iidd] + eta * (xx - w[iidd])
    return w
```
