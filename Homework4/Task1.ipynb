```python
# 动图生成类接口
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
 
class ImageSequenceGIF:
    def __init__(self, gif_dir='temp_gif_frames', gif_file='output.gif'):
        self.gif_dir = gif_dir
        self.gif_file = gif_file
        self.frames = []
        self.frame_count = 0
 
        if not os.path.exists(gif_dir):
            os.makedirs(gif_dir)
 
    def add_frame_from_plt(self, plt_figure=None):
        if plt_figure is None:
            plt_figure = plt.gcf()
 
        filename = os.path.join(self.gif_dir, f"frame_{self.frame_count:04d}.png")
        plt_figure.savefig(filename, bbox_inches='tight')
        self.frames.append(Image.open(filename))
        self.frame_count += 1
        plt.close(plt_figure)
 
    def save_gif(self, duration=100, last_frame_duration=200):
        if not self.frames:
            print("No frames to save.")
            return
 
        durations = [duration] * (len(self.frames) - 1) + [last_frame_duration]
        self.frames[0].save(
            self.gif_file,
            save_all=True,
            append_images=self.frames[1:],
            duration=durations,
            loop=0
        )
        print(f"GIF saved as {self.gif_file}")
 
    def clear_temp_frames(self):
        for f in os.listdir(self.gif_dir):
            os.remove(os.path.join(self.gif_dir, f))
        self.frames = []
        self.frame_count = 0
```

##### DNHH网络权系数建立并验证

```python
import numpy as np

def binarize(pattern):
    return np.where(pattern == 0, -1, 1)

def unbinarize(pattern):
    return np.where(pattern == -1, 0, 1)

def create_weight_matrix(patterns):
    n = patterns.shape[1]
    W = np.zeros((n, n))
    for p in patterns:
        W += np.outer(p, p)
    np.fill_diagonal(W, 0)  # 清零对角线
    return W

def update_state(W, state):
    return np.sign(W @ state)

def is_attractor(W, pattern):
    updated_state = update_state(W, pattern)
    return np.array_equal(updated_state, pattern)

letters_patterns = [
    letters_num[letter] for letter in letters_num 
]
binary_patterns = np.array([binarize(np.array(p)) for p in letters_patterns])
W = create_weight_matrix(binary_patterns)
print(W)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_matrix(matrix, title="Matrix Visualization"):
    matrix_array = np.array([[int(x) for x in row] for row in matrix])

    plt.figure(figsize=(10, 5))  
    plt.imshow(matrix_array, cmap='Greys', interpolation='nearest')  
    plt.title(title)  
    plt.axis('off')  
    plt.colorbar(label="Pixel Intensity") 
    plt.show()

visualize_matrix(W)
```

噪声添加与数据可视化

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def add_noise_to_pattern(pattern, noise_ratio=0.1):

    pattern = np.array(pattern) 
    noise_count = int(len(pattern) * noise_ratio)
    noise_positions = random.sample(range(len(pattern)), noise_count)
    noisy_pattern = pattern.copy()
    for pos in noise_positions:
        noisy_pattern[pos] = 1 - noisy_pattern[pos]
    return noisy_pattern

def show_character(pattern, offset_x=0, offset_y=0, color='black'):

    original_height = 12
    original_width = 10
    rows = [pattern[i * original_width:(i + 1) * original_width] for i in range(original_height)]
    
    for row_idx, row in enumerate(rows):
        for col_idx, pixel in enumerate(row):
            if pixel == 1:
                plt.scatter(offset_x + col_idx, offset_y + (11 - row_idx), s=50, color=color)
            else:
                plt.scatter(offset_x + col_idx, offset_y + (11 - row_idx), s=20, color='lightgrey')

def visualize_noisy_patterns(patterns_dict, gif_maker, colors, noise_ratio=0.1, frame_count=20):
    
    patterns = list(patterns_dict.values())
    labels = list(patterns_dict.keys())

    for frame_idx in range(frame_count):
        plt.clf()
        for idx, (pattern, label) in enumerate(zip(patterns, labels)):
            
            row = idx // 4
            col = idx % 4
            offset_x = col * 12
            offset_y = (1 - row) * 15

            noisy_pattern = add_noise_to_pattern(pattern, noise_ratio)
            show_character(noisy_pattern, offset_x, offset_y, colors[idx % len(colors)])
            plt.text(offset_x + 5, offset_y - 2, label, fontsize=12, ha='center')

        plt.axis('equal')
        plt.axis('off')
        plt.tight_layout()

        gif_maker.add_frame_from_plt(plt.gcf())

    gif_maker.save_gif(duration=100, last_frame_duration=500)
    gif_maker.clear_temp_frames()
```

DNHH网络与可视化

```python
def create_dhnn_weights(patterns):
    n = len(patterns[0])
    W = np.zeros((n, n))
    for pattern in patterns:
        p = np.array(pattern) * 2 - 1  # 转换为 {-1, 1}
        W += np.outer(p, p)
    np.fill_diagonal(W, 0)  # 清零对角线
    return W

def dhnn_update(weights, pattern):
    updated_pattern = np.dot(weights, pattern * 2 - 1)  # 转换为 {-1, 1} 更新
    return np.where(updated_pattern >= 0, 1, 0)  # 恢复到 {0, 1}

def visualize_dhnn_recall(patterns_dict, gif_maker, colors, noise_ratio=0.1, frame_count=20):
    
    patterns = list(patterns_dict.values())
    labels = list(patterns_dict.keys())
    weights = create_dhnn_weights(patterns)  # 使用存储模式创建权重矩阵

    noisy_patterns = [add_noise_to_pattern(pattern, noise_ratio) for pattern in patterns]
    
    for frame_idx in range(frame_count):
        plt.clf()

        for idx, (noisy_pattern, label) in enumerate(zip(noisy_patterns, labels)):
            row = idx // 4
            col = idx % 4
            offset_x = col * 12
            offset_y = (1 - row) * 15

            if frame_idx > 0:  # 从第二帧开始进行迭代更新
                noisy_patterns[idx] = dhnn_update(weights, noisy_patterns[idx])

            show_character(noisy_patterns[idx], offset_x, offset_y, colors[idx % len(colors)])
            plt.text(offset_x + 5, offset_y - 2, label, fontsize=12, ha='center')

        plt.axis('equal')
        plt.axis('off')
        plt.tight_layout()

        gif_maker.add_frame_from_plt(plt.gcf())

    gif_maker.save_gif(duration=100, last_frame_duration=500)
    gif_maker.clear_temp_frames()
```

```python
def dhnn_recall_multi_step(weights, pattern, max_steps=10):
    current_pattern = pattern.copy()
    for _ in range(max_steps):
        updated_pattern = dhnn_update(weights, current_pattern)
        if np.array_equal(updated_pattern, current_pattern):  # 检查是否已收敛
            break
        current_pattern = updated_pattern
    return current_pattern

def test_recovery_accuracy_multi_step(patterns, noise_ratio=0.1, max_steps=10, trials=5):

    weights = create_dhnn_weights(patterns)  
    correct_count = 0
    total_tests = len(patterns) * trials

    for pattern in patterns:
        for _ in range(trials):
            noisy_pattern = add_noise_to_pattern(pattern, noise_ratio) 
            recovered_pattern = dhnn_recall_multi_step(weights, noisy_pattern, max_steps)
            if np.array_equal(recovered_pattern, pattern):
                correct_count += 1

    return correct_count / total_tests  

noise_ratio = 0.1  
results = []

for num_patterns in range(1, len(letters) + 1):
    patterns_subset = list(letters.values())[:num_patterns]
    accuracy = test_recovery_accuracy_multi_step(patterns_subset, noise_ratio, max_steps=10, trials=10)
    results.append((num_patterns, accuracy))

num_patterns, accuracies = zip(*results)
plt.figure(figsize=(10, 6))
plt.plot(num_patterns, accuracies, marker='o', label=f'Noise Ratio = {noise_ratio}')
plt.title('Storage Capacity vs. Noise Resistance (Multi-step DHNN)')
plt.xlabel('Number of Stored Patterns')
plt.ylabel('Recovery Accuracy')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import copy

class BAM:
    def __init__(self):
        self.W = None

    def train(self, data_x, data_y):
        
        x_length = data_x[0].shape[0]
        y_length = data_y[0].shape[0]
        self.W = np.zeros((x_length, y_length))

        for x, y in zip(data_x, data_y):
            self.W += np.outer(x, y)
        print("权重矩阵 W 已计算。")

    def recall_forward(self, input_x, max_iterations=10):
        
        y = np.sign(self.W.T @ input_x)
        y[y == 0] = 1

        for _ in range(max_iterations):
            y_old = copy.deepcopy(y)
            x_new = np.sign(self.W @ y)
            x_new[x_new == 0] = 1
            y = np.sign(self.W.T @ x_new)
            y[y == 0] = 1
            if np.array_equal(y, y_old):
                break
        return y

    def recall_reverse(self, input_y, max_iterations=10):
        
        x = np.sign(self.W @ input_y)
        x[x == 0] = 1

        for _ in range(max_iterations):
            x_old = copy.deepcopy(x)
            y_new = np.sign(self.W.T @ x)
            y_new[y_new == 0] = 1 
            x = np.sign(self.W @ y_new)
            x[x == 0] = 1 
            if np.array_equal(x, x_old):
                break
        return x

def add_noise(vector, noise_level):
    noisy_vector = copy.deepcopy(vector)
    n_flip = int(len(vector) * noise_level)
    if n_flip == 0:
        return noisy_vector
    flip_indices = np.random.choice(len(vector), n_flip, replace=False)
    noisy_vector[flip_indices] *= -1
    return noisy_vector

# 数据集
letters_2 = {
    'G': G,  
    'N': N,
    'I': I,
    'Q': Q,
    'O': O,
    'U': U,
    'H': H,
    'Z': Z
}
data_x = [np.array(letters_2[letter]) for letter in letters_2.keys()]
data_y = [
    np.array([-1, -1, -1]),  # G
    np.array([-1, -1, 1]),   # N
    np.array([-1, 1, -1]),   # I
    np.array([-1, 1, 1]),    # Q
    np.array([1, -1, -1]),   # O
    np.array([1, -1, 1]),    # U
    np.array([1, 1, -1]),    # H
    np.array([1, 1, 1])      # Z
]

bam = BAM()
bam.train(data_x, data_y)


def test_bam(bam, letters_2, data_x, data_y, noise_levels=[0.1, 0.2]):

    for letter, x_original, y_original in zip(letters_2.keys(), data_x, data_y):
        print(f"\n=== 测试字母 '{letter}' ===")
        print(f"原始标签: {y_original}")

        for noise_level in noise_levels:
            # 添加噪声到 x
            noisy_x = add_noise(x_original, noise_level)
            recalled_y = bam.recall_forward(noisy_x)

        print(f"\n从标签回忆向量 '{letter}' 对应的 x")
        recalled_x = bam.recall_reverse(y_original)
        print(f"回忆向量 x: {recalled_x}")

test_bam(bam, letters_2, data_x, data_y)
```
